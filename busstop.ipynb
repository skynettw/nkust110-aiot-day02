```python
import requests
import json
url = 'https://data.coa.gov.tw/Service/OpenData/FromM/FarmTransData.aspx'
jsondata = requests.get(url).text
```

```python
data = json.loads(jsondata)
for item in data:
    try:
        if '高麗菜' in item['作物名稱']:
            print(item)
    except Exception as e:
        pass
print("done")
```

```python
url = 'https://www.nkust.edu.tw/p/403-1000-1363-{}.php' 
pages = [ url.format(page) for page in range(1, 49)]
print(pages)
```

```python
import requests, time, json
from bs4 import BeautifulSoup 
all_news_link = list()

for url in pages:
    try:
        html = requests.get(url).text 
        soup = BeautifulSoup(html, 'html.parser') 
        links = soup.find_all(class_="mtitle")
        for link in links:
            item = dict()
            item['title'] = link.a.text.strip()
            item['href'] = link.a['href']
            item['date'] = link.i.text
            all_news_link.append(item)
            print(item)
    except:
        pass
    time.sleep(3)
with open("nkustnews.json", "w", encoding="utf-8") as fp:
    json.dump(all_news_link, fp)
```
